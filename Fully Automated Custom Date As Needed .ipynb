```python
## Here task is to fetch dates startdate as 'last to last sunday' to 'last saturday'
## 0 (Monday) to 6 (Sunday)
import numpy as np 
import pandas as pd
import mysql.connector as mysql
from datetime import date
from datetime import timedelta

import datetime
def last_saturday(TodayDate):
    if TodayDate.weekday() <= 5:
        UpcomingSaturday = TodayDate + timedelta(days = (5 - TodayDate.weekday()))
    else:
        UpcomingSaturday = TodayDate + timedelta(days = 5)
    return UpcomingSaturday - timedelta(days = 7)

def last_to_last_sunday(TodayDate):
    if TodayDate.weekday() <= 6:
        UpcomingSaturday = TodayDate + timedelta(days = (6 - TodayDate.weekday()))
    else:
        UpcomingSaturday = TodayDate + timedelta(days = 6)
    return UpcomingSaturday - timedelta(days = 14)

enddate = last_saturday(date.today()).strftime('%Y-%m-%d')
startdate = last_to_last_sunday(date.today()).strftime('%Y-%m-%d')
```
